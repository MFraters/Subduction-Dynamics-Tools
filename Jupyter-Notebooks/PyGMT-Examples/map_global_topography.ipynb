```python
# Reads in GRD file of slabs location and depth for each slab from the Slab 2.0 data set

# Data was downloaded from: 
# Data is kept in ~/<Mybin/Data-Sets/Slab2Distribute_Mar2018

# Uses pygmt for plotting the depth as color
# on a global map (can be modified for preferred projection and region)

import pygmt
```

```python
pygmt.config(FONT='Times-Roman')
pygmt.config(FONT_LABEL='Times-Roman,12p')

print('To see Colormap go to: https://docs.generic-mapping-tools.org/6.4/cookbook/cpts.html')
print('Some good topography color maps are:')
print('section: gmt')
print('gray, earth, topo, topo1')
print('geo, terra, world, relief')
print('section: SCM (scientific color maps)')
print('bukavu, fes, oleron')
colormap = input('Which color map do you want to use? ')
```

```python
# Basic global map in gmt
# Edit projection to change which type of global projection
# Edit 180 to change longitude of center of map
fig = pygmt.Figure()
fig.basemap(region="d", projection="W180/15c", frame=True) 
grid = pygmt.datasets.load_earth_relief(resolution="10m",registration="gridline")
# Use this method so the colorscale is equalized to the area of the plot. 
colormap1 = pygmt.grd2cpt(grid=grid,cmap=colormap)
fig.grdimage(grid=grid,cmap=colormap1)
fig.coast(shorelines="1/0.5p,black")
fig.colorbar(cmap=colormap1,position="JMR", box=False, frame=["x+lTopography", "y+lm"])
fig.show()

pngfile = 'map_global_bathymetry_' + str(colormap) + '.png'
# Save figure to file
print('Saving to file ', pngfile)
fig.savefig(pngfile)
```
