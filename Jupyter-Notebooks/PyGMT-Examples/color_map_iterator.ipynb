```python
# Example of how to iterate through colors pulled from a color map
import numpy as np
from matplotlib.pyplot import cm
```

```python
n = 25
color = iter(cm.rainbow(np.linspace(0, 1, n)))

for i in range(n):
    c = next(color)
    rgb = str(int(255*c[0])) + '/' + str(int(255*c[1])) + '/' + str(int(255*c[2]))
    pen1 = '2p,' + rgb  # for use in pygmt
    print(c, '-->', pen1)
```

```python
# Example just printing the color-map
print(cm.Blues(np.linspace(0, 1, n)))
```
